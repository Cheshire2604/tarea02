```python
#TAREA
import numpy as np
import cv2
#LOOPEA EL VIDEO
cap = cv2.VideoCapture('solidWhiteRight.mp4')
# Capture frame-by-frame
if (cap.isOpened()==False):
 print("error abriendo video")

while (True):
        #leer frame por frame
    ret, image = cap.read()
    if ret==True: 
        #Convertir a la escala de grises
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Filtro Gaussiano
        kernel_size = 9;
        blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size),0)

     # Detección de bordes
        low_threshold = 50
        high_threshold = 150
        edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
    
    # Next we'll create a masked edges image using cv2.fillPoly()
        mask = np.zeros_like(edges)   
        ignore_mask_color = 255   

    # This time we are defining a four sided polygon to mask
        imshape = image.shape
        vertices = np.array([[(0,imshape[0]),(imshape[1]*3/7, imshape[0]*9/16), (imshape[1]*4/7, imshape[0]*9/16), (imshape[1],imshape[0])]], dtype=np.int32)
        cv2.fillPoly(mask, vertices, ignore_mask_color)
        edges = cv2.bitwise_and(edges, mask)
    
    # Define the Hough transform parameters for large lines

    # Make a blank the same size as our image to draw on
        rho = 2 # distance resolution in pixels of the Hough grid
        theta = np.pi/180 # angular resolution in radians of the Hough grid
        threshold = 70    # minimum number of votes (intersections in Hough grid cell)
        min_line_length = 170 #minimum number of pixels making up a line
        max_line_gap = 150    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
        lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

    # Dibujar las lineas detectadas
        image2show = np.copy(image)

        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(image2show,(x1,y1),(x2,y2),(244,0,161),3)        
        
    # Display the resulting frame
        cv2.imshow('frames',image2show)
        if cv2.waitKey(1) & 0xFF == ord('q'): 
            break
    else:
        cap=cv2.VideoCapture('solidWhiteRight.mp4')
# When everything done, release the capture



cap.release()
# Closes all the frames
cv2.destroyAllWindows()
    
```
